```python
import tensorflow as tf
from tensorflow import keras
```

```python
## Functional API

## 1번 세션 클리어
keras.backend.clear_session()
## 2번 레이어 사슬처럼 엮기
il = keras.layers.Input(shape=(5,5,3))
cl = keras.layers.Conv2D(filters=1, # 제작하려는 feature map의 수
                         kernel_size=(3,3), # fileter의 가로세로 사이즈
                         strides=(1,1), # filter가 어떻게 훑을 것인지 값 지정
                         activation='relu')(il)

pl = keras.layers.MaxPool2D(pool_size=(2,2), strides=(2,2), padding='same')
## 3번 모델의 시작과 끝 지정
model = keras.models.Model(il,cl)

model.summary()
```


- strides 2,2 가 적용이 되려면 패딩이 없었을 땐 존재하는 칸 안에서 한 번만 작동하는데
- 패딩이 들어가는 순간 영역이 넓어져서
