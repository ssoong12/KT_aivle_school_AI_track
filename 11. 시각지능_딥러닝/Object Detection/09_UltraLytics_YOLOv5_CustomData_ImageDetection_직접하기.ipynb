```python
## yolov5 폴터 requirements.txt 수정 필요
## setuptools<=64.0.2

temp_str = 'setuptools<=64.0.2\n' 

f = open('/content/yolov5/requirements.txt', 'r') 
f_str = f.readlines() 
f.close() 

f2 = open('/content/yolov5/requirements.txt', 'w') 

for idx, val in enumerate(f_str) : 
    if 'setuptools' in val : 
        idx_v = idx 
        f_str.remove(val) 
        f_str.insert(idx_v, temp_str) 
        
for val in f_str : 
    f2.write(val) 
    
f2.close()
```

### 3) train.py 실행

- 명령어 도움말 : python train.py -h

```python
# dataset의 위치와 클래스에 대한 정보
# 모델 구조에 대한 설명도 얌으로 되어 있음
# weigths는 다운로드 받은거 그냥 넣으셈
# 에폭스 최대 10번
# patience 3번만 참을게요~

!cd yolov5; python train.py \
    --data '/content/drive/MyDrive/my_data/datasets/금발_흑발_민머리.yaml' \
    --cfg '/content/yolov5/models/yolov5s.yaml' \
    --weights '/content/yolov5/pretrained/yolov5s.pt' \
    --epochs 10 \
    --patience 3 \
    --img 640 \
    --project 'trained' \
    --name 'train_street' \
    --exist-ok
    # --device cpu
```
