## 데이터 전처리

```python
def target_morphs(data, target = None, target_dynamic = None):
    df = data.copy()
    # target이 지정되지 않았으면 가장 마지막 열로 지정
    if target is None:
        target = data.columns[-1]
    # targets가 존재하지만 리스트가 아니라면 리스트로 변경
    if target_dynamic is not None:
        if not isinstance(target_dynamic, list):
            target_dynamic = [target_dynamic]
        # target이 targets에 포함되는지 여부를 target에 저장
        df[target] = df[target].apply(lambda x: 0 if x in target_dynamic else 1)
    return df
```

```python
from sklearn.metrics import classification_report, confusion_matrix

def score_model(data, random_state = None):
    X, y, lb = data_processing(data)
    x_train, x_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = random_state)
    print('='*150)
    y_pred = double_model(x_train, y_train, x_test)
    print(classification_report(y_test['Activity_dynamic'], y_pred))
    print('='*150)
    # 썼던 라벨(lb)가 주어진다면
    if lb is not None:
        # LabelEncoder가 숫자형 범주로 바꾸는데 사용한 class들을 가져오고
        values = lb.classes_
        # 그걸 다시 LabelEncoder를 이용해 숫자로 바꿔서
        keys = lb.transform(lb.classes_)
        # 재결합해서 번역 Dict 생성
        dictionary = dict(zip(keys, values))
        for _ in range(len(dictionary)):
            # 그래서 Confusion Matrix를 보기 쉽게 출력
            print(dictionary[_], end=', ')
    print()
    print(confusion_matrix(y_test['Activity_dynamic'], y_pred))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

def data_processing(data, target = 'Activity'):
    # target_dynamic 리스트에 들어있는 Activity는 0, 없으면 1
    df = target_morphs(data, target = target, target_dynamic = ['STANDING', 'LAYING', ' SITTING'])
    # Unused
    # ynum = len(lb.classes_)
    # 위의 is dynamic이 아닌 실제 값. morphs는 Activity를 덮어씌우게 만들어서 이렇게 따로 추가해 줘야 함.
    real_y = data['Activity_dynamic'].copy()
    # Activity_dynamic 에 저장
    df[f'{target}_dynamic'] = real_y
    target = [target, f'{target}_dynamic']
    X = df.drop(target, axis = 1)
    y = df[target]
    # lb를 반환하는 이유는 score_model에서
    return X, y, lb
X, y, lb = data_processing(data)
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 2023)
```

```python
def double_model(x_train, y_train, data):
    # 1차 분류
    model1 = RandomForestClassifier()
    model1.fit(x_train, y_train.drop('Activity_dynamic', axis=1))
    # 분류 예측 결과
    y1_pred = model1.predict(x_train)
    # 0이면 여기에
    x_train_a = x_train.iloc[y1_pred == 0]
    # 1이면 여기에
    x_train_b = x_train.iloc[y1_pred == 1]
    # 0 -> 0, 1, 2 분류
    model2 = RandomForestClassifier()
    model2.fit(x_train_a, y_train.drop('Activity', axis=1).iloc[y1_pred == 0])
    # 1 -> 3, 4, 5 분류
    model3 = RandomForestClassifier()
    model3.fit(x_train_b, y_train.drop('Activity', axis=1).iloc[y1_pred == 1])

    temp = model1.predict(data)
    # 원본 순서를 저장
    idx = data.index
    # 데이터 분리
    data_a = data.iloc[temp == 0]
    data_b = data.iloc[temp == 1]
    ans_a = pd.DataFrame({'answer':model2.predict(data_a)}, index = data_a.index)
    ans_b = pd.DataFrame({'answer':model3.predict(data_b)}, index = data_b.index)
    # 결과 합치기
    ans = pd.concat([ans_a, ans_b])
    # 들어온 데이터 순서대로 재정렬해서 반환
    return ans.loc[idx]
```

## 전처리 수정

```python
def select_top_features(model, x_train, x_test, num_features=30):
    """
    Select top features based on feature importance of model
    """
    # 모델 학습
    model.fit(x_train, y_train)

    # 변수 중요도 계산
    feature_importances = model.feature_importances_

    # 상위 num_features개의 변수 추출
    top_features = x_train.columns[np.argsort(feature_importances)[::-1][:num_features]]

    # 새로운 데이터프레임 생성
    x_train_new = x_train[top_features]
    x_test_new = x_test[top_features]

    return x_train_new, x_test_new
```

### CatBoost 튜닝

```python
from sklearn.model_selection import GridSearchCV
from catboost import CatBoostClassifier

CBmodel = CatBoostClassifier()

param_grid = {
    'learning_rate': [0.05, 0.1, 0.2],
    'depth': [4, 6, 8],
    'l2_leaf_reg': [1, 3, 5],
    'iterations': [500, 1000]
}

grid_search = GridSearchCV(
    estimator=CBmodel, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)

grid_search.fit(x_train, y_train.drop('Activity', axis=1))

best_params = grid_search.best_params_
print(f"Best parameters: {best_params}")

CBmodel = CatBoostClassifier(**best_params)
CBmodel.fit(x_train, y_train.drop('Activity', axis=1), verbose=False)

CBmodel.score(x_test, y_test.drop('Activity', axis=1))
y_pred = CBmodel.predict(test)
print(classification_report(y_test['Activity_dynamic'], y_pred))
print(confusion_matrix(y_test['Activity_dynamic'], y_pred))
```
