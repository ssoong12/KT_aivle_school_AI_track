# 다룰 내용

- 학습 데이터 양이 많으면 성능이 좋아질까요?
- 학습 데이터 양에 따른 성능(Learning Curve)을 확인해봅니다.

```python
# 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings(action='ignore')
%config InlineBackend.figure_format='retina'
```

```python
# 데이터 읽어오기
path = 'https://raw.githubusercontent.com/jangrae/csv/master/airline_satisfaction.csv'
data = pd.read_csv(path)

# 10,000개만 추출
data = data.head(10000) 

# 확인
data.head()
# 만족합니다, 만족하지 않습니다 > 분류 문제
```

```python
# 불필요한 변수 제거
del_cols = ['id', 'departure/arrival_time_convenient', 'gate_location', 'departure_delay_in_minutes']
data.drop(del_cols, axis=1, inplace=True)

# 결측치 제거
data.dropna(inplace=True)

# 가변수화(범주형 데이터는 가변수화를 시켜야하고, 다중공선성 해결을 위해 첫번째거는 날려버림)
dumm_cols = ['gender', 'customer_type', 'type_of_travel', 'class']
data = pd.get_dummies(data, columns=dumm_cols, drop_first=True)

# 확인
data.head()
```

```python
# x, y 분리
target = 'satisfaction'

x = data.drop(target, axis=1)
y = data.loc[:, target]

# 데이터 크기 확인
x.shape
```

**1) max_depth=3**

```python
# 모듈 둘러오기
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import learning_curve

# 모셀선언
model = DecisionTreeClassifier(max_depth=3)

# Learning Curve 수행
tr_size, tr_scores, val_scores = learning_curve(model, 
                                                x, 
                                                y,
                                                train_sizes=range(10, 7900, 20),
                                                shuffle=True,
                                                cv=5) # range안에 숫자는 10부터 7900까지 20 단위로 10, 30, 50, 70...
```

```python
# CV 결과 --> 평균
val_scores_3_mean = val_scores.mean(axis=1)

# 시각화
plt.figure(figsize=(8, 5))
plt.plot(tr_size, val_scores_3_mean)
plt.title('Learning Curve', size=20, pad=15)
plt.ylabel('Score')
plt.xlabel('Training Size')
plt.show()
# 데이터가 많다고 해서 성능이 올라가는 것은 아니다. 이런 만족도 조사 같은 경우에는 1000건 정도만 있어도 된다. 적당한 양이 중요한 것, 천만 건이 의미가 없다
```

**2) max_depth=5**

```python
# 모듈 둘러오기
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import learning_curve

# 모셀선언
model = DecisionTreeClassifier(max_depth=5)

# Learning Curve 수행
tr_size, tr_scores, val_scores = learning_curve(model, 
                                                x, 
                                                y,
                                                train_sizes=range(10, 7900, 20),
                                                shuffle=True,
                                                cv=5) # 값이 10, 30, 50... 이렇게 뽑혔다면 500번을 곱한다 cv = 5
```

```python
# CV 결과 --> 평균
val_scores_5_mean = val_scores.mean(axis=1)

# 시각화
plt.figure(figsize=(8, 5))
plt.plot(tr_size, val_scores_3_mean)
plt.plot(tr_size, val_scores_5_mean)
plt.title('Learning Curve', size=20, pad=15)
plt.ylabel('Score')
plt.xlabel('Training Size')
plt.legend(['max_depth=3', 'max_depth=5'])
plt.show()
# 많은 양의 데이터가 중요하지 않다. 질이 중요한 거다. 좋은 데이터가 있으면 조금만 있어도 된다.
# 신뢰성 없는 문제집 1000권 만권 보는 것보다, 신뢰성 있는 문제집 1권만 있으면 된다.
# max_deth로 성능 조정이 된다.
# 많은 양의 데이터를 위해 고생하지 않아도 된다.
```
